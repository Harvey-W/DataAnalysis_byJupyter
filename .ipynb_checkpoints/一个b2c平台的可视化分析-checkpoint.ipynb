## chapter 1 品牌和行业综述

```python
try:
    conn = mysql.connect(host='192.***.*.***', port=****, user='analyst_user', \
                           passwd='234joiU#%hj6kh*#5HFk@$35',db='jmw_message',charset='utf8')
    cur = conn.cursor()
    sql_brand = 'SELECT v1.sname as 一级行业,v2.sname as 二级行业, p.brand_name as 品牌, COUNT(DISTINCT(t.telphone)) as 留言数 FROM t_message as t \
    LEFT JOIN jmw_company.project as p ON p.id=t.target_id \
    LEFT JOIN jmw_company.v9_conv_type as v1 ON v1.id=p.industry_id \
    LEFT JOIN jmw_company.v9_conv_type as v2 ON v2.id=p.industry_child_id \
    WHERE publish_date between 20171001 and 20171031 \
    and origin not like "4%" \
    GROUP BY 品牌 \
    ORDER BY 留言数 DESC'
    df1 = pd.read_sql(sql_brand, conn)
except Exception as e:
    df1 = pd.read_excel('msgs.xlsx',sheetname='brand')
```

```python
df1.insert(0,'id',[id for id in range(1,len(df1.品牌)+1)])
```

```python
df1.head()
```

```python
df1.info()
```

```python
user_total = df1.groupby('id').留言数.sum().sort_values().reset_index()
user_total_cumsum = user_total.cumsum()

cumsum_max = user_total_cumsum.max()
user_value_rate = user_total_cumsum.apply(lambda x:x.留言数/cumsum_max, axis=1)
user_value_rate.留言数.plot(figsize=(8,4), label='品牌价值图-留言数')
plt.legend()
```

> **83%的品牌贡献了30%的留言**

## chapter 2 数据清洗——个体异常值

```python
conn = mysql.connect(host='192.168.8.172', port=3306, user='analyst_user', \
                       passwd='234joiU#%hj6kh*#5HFk@$35',db='jmw_message',charset='utf8')
cur = conn.cursor()
sql_3 = 'SELECT t.message_user_id as user_id,t.publish_date as raw_date,v.sname as industry,p.brand_name as brand, COUNT(DISTINCT t.id) as message \
FROM t_message as t \
LEFT JOIN jmw_company.project as p ON p.id=t.target_id \
LEFT JOIN jmw_company.v9_conv_type as v ON v.id=p.industry_id \
WHERE publish_date between 20171001 and 20171031 \
and origin not like "4%" \
GROUP BY user_id,p.brand_name \
ORDER BY publish_date'
df3 = pd.read_sql(sql_3, conn)
```

```python
df3['date'] = pd.to_datetime(df3.raw_date, format='%Y%m%d')
```

```python
df3.head()
```

```python
plt.figure(figsize=(15,4))
plt.subplot(121)
ax1 =df3.groupby('date').user_id.count().plot(label='by date')
plt.legend(loc=4)

plt.subplot(122)
ax2 =df3.groupby('date').message.sum().plot(label='by user')
plt.legend(loc=4) #bbox_to_anchor=[0.5, 1]
```

```python
x1 = df3.groupby('industry').brand.count()
y1 = df3.groupby('industry').message.sum()
scatter1 = pd.DataFrame(data=[x1,y1]).T
scatter1 = scatter1.drop(labels='餐饮') #去掉极值
scatter1.plot.scatter(x='brand',y='message',label='scatter of per brand')
plt.legend(loc=0)
#品牌数越多，平均每个品牌带来的留言数越多
```

```python
x2 = df3.groupby('date').user_id.count()
y2 = df3.groupby('date').message.sum()
scatter2 = pd.DataFrame(data=[x2,y2]).T
scatter2.plot.scatter(x='user_id',y='message', label='scatter of per user')
plt.legend(loc=0)
```

```python
df3.groupby('user_id').message.sum().hist(bins=30, label='frequency of users')
plt.legend()
```

> **个体异常值影响较小**

# chapter 3 头部品牌

- 品牌分析
    - 构成
    - 发源地
    - 口碑

```python
head_brand=df1.iloc[2:2078]
```

```python
head_brand.head()
```

```python
head_brand_sorted = pd.DataFrame(head_brand.groupby(['一级行业','二级行业']).留言数.sum().sort_values(ascending=False))
```

```python
head_brand_sorted['品牌数'] = head_brand.groupby(['一级行业','二级行业']).品牌.count()
```

```python
head_brand_sorted['平均留言数'] = head_brand_sorted.留言数/head_brand_sorted.品牌数
head_brand_reset = head_brand_sorted.reset_index()
head_brand_sorted.head()
```

> 行业价值 = 平均带来留言数

```python
head_brand_reset.平均留言数.hist(bins=10, label='行业价值分布')
plt.legend()
```

```python
plt.hist(head_brand_reset[head_brand_reset.一级行业=='餐饮'].平均留言数,5, normed=1,facecolor='green',alpha=0.5, label='餐饮')
plt.hist(head_brand_reset[head_brand_reset.一级行业=='服装'].平均留言数,5, normed=1,facecolor='red',alpha=0.5, label='服装')
plt.hist(head_brand_reset[head_brand_reset.一级行业=='教育'].平均留言数,5, normed=1,facecolor='blue',alpha=0.5, label='教育')
plt.legend()
```

```python
head_brand_sorted.boxplot(column='平均留言数', by='一级行业',figsize=(15,10))
```

>- 朝阳行业：新行业、珠宝、干洗、环保
- 衰退行业：餐饮、教育、零售

```python
head_brand_sorted.groupby('一级行业').品牌数.sum().sort_values().plot.bar(figsize=(15,10), label='行业品牌数')
plt.legend(loc=2,bbox_to_anchor=(0,1))
```

> **传统优势行业（项目数前五名/占比64%的行业）性价比降低，新兴行业崛起明显**

## chapter 5 竞争对手

```python
df4 = pd.read_excel('data.xlsx',sheetname='Sheet2')
df4['per_month']=df4.per_month.round(decimals=0)
```

```python
df4_clean = df4.iloc[0:296] #每月留言多于1
df4_clean.insert(0,'id',[id for id in range(1,len(df4_clean.per_month)+1)])
```

```python
df4_clean.head()
```

```python
user_total = df4_clean.groupby('id').per_month.sum().sort_values().reset_index()
user_total_cumsum = user_total.cumsum()

cumsum_max = user_total_cumsum.max()
user_value_rate = user_total_cumsum.apply(lambda x:x.per_month/cumsum_max, axis=1)
user_value_rate.per_month.plot(figsize=(8,4), label='竞争对手品牌价值图-留言数')
plt.legend()
```

> 84%的品牌带来了20%的留言（反证留言取值正确的可能性）
